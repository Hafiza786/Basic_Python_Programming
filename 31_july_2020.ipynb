# Basic 2D Plotting Functions

```python
# plot linear graph
import matplotlib.pyplot as plt
```

```python
#linear graph
plt.title("Basic linear graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
#Non linear graph
#plot a graph between distance and time
Distance=list(range(2,11,2)) #y-axis
Time=list(range(1,6))
plt.title("Distance vs Time")
plt.xlabel ("Time")
plt.ylabel ("Distance")
plt.plot(Time,Distance,'yo')#yo colour code yellow dots
plt.grid()
plt.show()
```

```python
#Non linear graph
#plot a graph between distance and time
Distance=list(range(2,11,2)) #y-axis
Time=list(range(1,6))
plt.title("Distance vs Time")
plt.xlabel (input("enter"))
plt.ylabel (input("enter"))
plt.plot(Time,Distance,'yo')#yo colour code yellow dots
plt.grid()
plt.show()
```

```python
# Non linear graph   for the equation v=a*t
#time vs voltage
t= list(range(1,11))
a =list(range(1,len(t)+1))
v =[a[i]*i for i in range (len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
#voltage vs time
import numpy as np
t=np.arange(0.1,2.0,0.1)# range from 0.1 50 2.0 with difference btw values as 0.1
v=[1+np.sin(2*np.pi*i) for i in t] # v=1+sin(2pit)
plt.plot(t,v)
plt.grid()
plt.xlabel("Time")
plt.ylabel("voltage")
plt.title("Time vs Voltage")
plt.show()
```

# Line Plot

```python
x  = np.arange(10)
y  = x**2
plt.plot(x,y)
plt.show()
```

```python
x  = np.arange(10)
y  = x**2
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(" Y = X**2 ")
plt.plot(x,y)
plt.show()
```

```python
x  = np.arange(10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(" Line Plot ")
plt.plot(x,x,label = "Line Plot")
plt.plot(x,x**2,label = "Square Plot")
plt.plot(x,x**3,label = "Cube plot")
plt.legend()
plt.show()
```

```python
x  = np.arange(10)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(" Line Plot ")
plt.plot(x,x,label = "Line Plot",linestyle = "--")
plt.plot(x,x**2,label = "Square Plot",linestyle= "-.")
plt.plot(x,x**3,label = "Cube plot",linestyle=":")
plt.legend()
plt.show()
```

# Bar

```python
x  = np.arange(1,100,2)
y  = np.linspace(1,50,len(x))
print(max(x),max(y))
print(min(x),min(y))
plt.xlabel("distance in m")
plt.ylabel("speed in km/hr")
plt.xlim(40,80)
plt.ylim(20,70)
plt.bar(x,y)
plt.show()
```

```python
x = ["Firoz","Afsana","Althaf","Abdul"]
y = [90,97,87,95]
plt.bar(x,y)
plt.xlabel("Name of Students")
plt.ylabel("Python Marks")
plt.title("Progress")
plt.show()
```

# Pie Chart

```python
x  = [90,98,78,67,56,78]
plt.pie(x,labels = ["Py","c","c++","java","R","embedded C"],colors=["g","b","m","red","c","gray"])
plt.legend()
plt.xlim(3,3)
plt.show()
```
